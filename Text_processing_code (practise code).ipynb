```python
### 3.NLP Pipeline
```

```python
#Lets cover all the preprocessing steps into a single function now
```

```python
def lower_replace(series):
    output = series.str.lower()
    output = output.str.replace(r'\[.*?\]', '', regex=True)
    output = output.str.replace(r'[^\w\s]', '', regex=True)
    return output

def token_lemma_stopwords(text):
    doc = nlp(text)
    output = [token.lemma_ for token in doc if not token.is_stop]
    output = ' '.join(output)
    return output
    
def filtering_on_pos(text, pos_list= ['NOUN', 'PROPN']):
    doc = nlp(text)
    output = [token.text for token in doc if token.pos_ in pos_list]
    output = ' '.join(output)
    return output
```

```python
#now what we are going to do is combine all these functions within another function, and because we are using the above fucntions as part of
#another function, the above functions are called as HELPER FUNCTIONS
```

```python
def nlp_pipeline(series):
    output = lower_replace(series)
    output = output.apply(token_lemma_stopwords)
    output = output.apply(filtering_on_pos)
    return output
```

```python
test_series
```

```python
df.sentence
```

```python
nlp_pipeline(test_series)
```

```python
nlp_pipeline(df.sentence)
```

```python
#before working with count vectorization, if you open the above code next day, youll have to rerun the whole thing again, instead of that, just open the pickle file you have saved above.
```

```python
import pandas as pd
series = pd.read_pickle('text_clean.pkl')
series
```

```python
from sklearn.feature_extraction.text import CountVectorizer

#create a basic count vectorizer object

cv = CountVectorizer()

#the above cv is basic, but we will add other features next, cv=CountVectorizer(stop_words='english', ngram_range=(1,2), min_df=2)
# We are doing the stop words here again. It is your choice to wanna do it in Spacy or in sk-learn, but we can do stop words here as well.
#then we are doing ngram_range, (1,1)  is default, likely gives "data", then if you use (1,2) likely gives "data"," data science", then if we use (3,3) like gives "data science workflow" 
#then finally we have min_df, which says minimum document frequency, it only returns the words that are in a minimum of 'n' documents, n being a number or percentages.

#after creating a CV object, you take that, and you do fit_transform of your series
#what fit_transform does here is it divides into two parts, what "fit" does is it recognises all your text, it tokenizes it, and it finds all the unique terms in the text, and makes those terms into your columns data 
#and then what the "transform" does is, it looks at every row in your data frame, and it builds in the words count below, so it actually shows how often each term appears in each document
dtm = cv.fit_transform(series)
#toarray() is used to get the DTM, if we dont do its just a compressed sparce matrix
#coulums=cv.get_feature_names_out is to get the column names, if we dont use it, there will just be 1s and 0s.
dtm_df = pd.DataFrame(dtm.toarray(), columns=cv.get_feature_names_out())
dtm_df
```

```python
cv2 = CountVectorizer(stop_words='english', ngram_range=(1,2), min_df=.2, max_df=.8)
dtm2 = cv2.fit_transform(series)
dtm_df2 = pd.DataFrame(dtm2.toarray(), columns=cv2.get_feature_names_out())
dtm_df2
```

```python
# Let's say your business asked what is the most appeared word in this doc
term_frequency = dtm_df2.sum()
term_frequency
```

```python
term_frequency.sort_values().plot(kind='barh');
```
